```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df = pd.read_csv('../data/CCDC_descriptors.csv', delimiter=';', decimal=',')
```

```python
# count the number of samples belonging to the class of presence/absence of the target property
property_list = [['Unobstructed', (df['Unobstructed']==1).sum(), 'presence'], 
                ['Orthogonal planes', (df['Orthogonal planes']==1).sum(), 'presence'], 
                ['H-bonds bridging', (df['H-bonds bridging']==1).sum(), 'presence'],
                ['Unobstructed', (df['Unobstructed']==0).sum(), 'absence'], 
                ['Orthogonal planes', (df['Orthogonal planes']==0).sum(), 'absence'], 
                ['H-bonds bridging', (df['H-bonds bridging']==0).sum(), 'absence']]
```

```python
# transfer data to the dataframe
df_property = pd.DataFrame(property_list, columns=['Mechanical property', 'Number of samples', 'Label'])
df_property
```

```python
# plot the class representation for each target property
f, ax = plt.subplots(figsize=(5, 2.5), dpi=200)

sns.set_palette('mako', n_colors=2)
sns.barplot(data=df_property, x="Mechanical property", y="Number of samples", hue="Label", width=0.6,
    linewidth=2, edgecolor="1.0")

ax.set(ylim=(0, 5500))
sns.despine(right=True)

ax.spines['left'].set_linewidth(1)
ax.spines['left'].set_color('#CCCCCC')

ax.spines['bottom'].set_linewidth(2)
ax.spines['bottom'].set_color('#CCCCCC')

ax.tick_params(left=False, bottom=False)

ax.set_xticklabels(labels = [
            "Unobstructed"  + "\n" + "planes",
            "Orthogonal"  + "\n" + "planes",
            "H-bonds"  + "\n" + "bridging"])
```
